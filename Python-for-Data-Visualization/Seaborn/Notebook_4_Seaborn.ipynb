```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False, bins=25)
```

```python
len(tips.columns)
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.pairplot(tips,hue='sex')
```

```python
tips
```

```python
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',estimator=np.std, hue='smoker')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill', hue='smoker')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
flights=sns.load_dataset('flights')
```

```python
fp=flights.pivot_table(index='month', columns='year',values='passengers')
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,row='smoker',col='time')
g.map(sns.distplot,'total_bill')
```

```python
tips=sns.load_dataset('tips')
sns.lmplot(data=tips,x='total_bill',y='tip',col='day', hue='sex')
```

```python
sns.set_context('talk', font_scale=1.15)
sns.set_style('whitegrid')
sns.countplot(data=tips,x='sex')
```
