```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r-')

plt.subplot(1,3,2)
plt.plot(y,x,'b')

plt.subplot(1,3,3)
plt.plot(x,y,'r-')
```

```python
fig=plt.figure()
```

```python
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
axes.plot(x,y)
```

```python
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('OO Main Title')
```

```python
axes.plot(x,y)
```

```python
fig
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title('Bigger Plot')
axes2.set_title('Smaller Plot')

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()

axes[0][1].plot(x,y)
axes[1][1].plot(-y,x)
axes[1][0].plot(y,x)
axes[0][0].plot(-x,y)
```

```python
fig=plt.figure(figsize=(3,2),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(5,4),nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout
```

```python
fig.savefig('my_picture.png', dpi=100)
```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x**2',marker='o')
axes.plot(x,x**3,label='x**3', marker='*')
axes.set_xlim([-2,6])
axes.set_ylim([4,30])
axes.legend(loc=10)
```
